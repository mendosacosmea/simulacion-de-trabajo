# Ingesta de Datos y Configuración del Entorno
**Primer paso** Recepción del legado de datos de los departamentos de la empresa y migración a un entorno de trabajo relacional (SQLite).

### 1. Recepción de Información (RAW Data)
He recibido los archivos en formato CSV correspondientes a las áreas operativas. 
El estado actual de los datos es "crudo", lo que significa que presentan inconsistencias que deben ser tratadas antes de cualquier análisis:

- `RAW_EMPLEADOS.csv`: Estructura de personal.
- `RAW_VENTAS.csv`: Registros transaccionales.
- `RAW_COMPRAS.csv`: Registro de proveedores y adquisiciones.
- `RAW_COMPRA_MATERIA_PRIMA.csv`: Registro de materia prima para la produccion.

## Segundo paso. Preparación del Entorno
Para garantizar la integridad y escalabilidad, he decidido migrar estos archivos de un formato de texto plano (CSV) a una base de datos relacional **SQLite**. 

**Ventajas de esta migración:**
- Definición estricta de tipos de datos (INTEGER, REAL, DATE).
- Capacidad de ejecutar consultas complejas para validación.
- Seguridad al realizar operaciones de limpieza sin alterar los archivos originales (RAW).

## Tercer paso. Limpieza de los Datos
Una de las primeras tareas es la limpieza propiamente dicha de los datos crudos, aca es donde nos encontramos con las inconsitencias de typo de datos de las
diferentes columnnas que en muchas ocaciones no aparcen como nosotros desemamos, entonces empieza el trabajo sucio, la limpieza de la tabla.

**Objetivo:**
Comenzamos con la tabla *Ventas* en donde nos dirigimos a la campo *Cantidad* y podemos notar que hay inconsistencias en los registros (tipos de datos) que tiene,
necesitamos un tipo de dato numérico en esa tabla y vemos que hay palabras en lugar de numero, lo que vamos a realizar es que esas palabras se conviertan en 
números y aquellas celdas que no podeean dato (null), contengan el cero.
